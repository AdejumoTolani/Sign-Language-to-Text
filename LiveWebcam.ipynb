```python
import tensorflow as tf
import cv2
import os
import numpy as np
from tensorflow import keras
from cvzone.HandTrackingModule import HandDetector
import math
```

```python
model = keras.models.load_model('my_model_five.keras')
```

```python
cap = cv2.VideoCapture(0)
```

```python
detector = HandDetector(maxHands=2)
```

```python

spacing = 20
# to give our hands an offset at the sides to not miss any movement
imgWhiteSize = 64

folder = {0: 'A',
 1:'B',
 2:'C',
 3:'D',
 4:'Delete',
 5:'E',
 6:'F',
 7:'G',
 8:'H',
 9:'I',
 10:'J',
 11:'K',
 12:'L',
 13:'M',
 14:'N',
 15:'O',
 16:'P',
 17:'Q',
 18:'R',
 19:'S',
 20:'Space',
 21:'T',
 22:'U',
 23:'V',
 24:'W',
 25:'X',
 26:'Y',
 27:'Z'}


while True:
    success, img = cap.read()  
    ImageShow = img
    hands, image2 = detector.findHands(img)
    
    if hands:
        hand = hands[0]
        x, y, w, h = hand['bbox']
        imgWhite = np.ones((imgWhiteSize, imgWhiteSize, 3), np.uint8) * 250
        imgCrop = img[y-spacing:y + h + spacing, x - spacing:x + w + spacing]
        
        imgCropShape = imgCrop.shape
        aspectRatio = h/w

        if aspectRatio > 1:
            k = imgWhiteSize / h
            wCalculated = math.ceil(k*w)
            imgResize = cv2.resize(imgCrop, (wCalculated,imgWhiteSize))
            img = cv2.cvtColor(imgResize,cv2.COLOR_BGR2GRAY)
            imgResizeShape = imgResize.shape
            widthGap = math.ceil((imgWhiteSize-wCalculated)/2)
            imgWhite[:, widthGap:wCalculated+widthGap] = imgResize



        else:
            k = imgWhiteSize / w
            hCalculated = math.ceil(k*h)
            imgResize = cv2.resize(imgCrop, (imgWhiteSize,hCalculated))
            img = cv2.cvtColor(imgResize,cv2.COLOR_BGR2GRAY)
            imgResizeShape = imgResize.shape
            heightGap = math.ceil((imgWhiteSize-hCalculated)/2)
            imgWhite[heightGap:hCalculated + heightGap, :] = imgResize


        cv2.imshow("ImgWhite",imgWhite)
        imgWhite = cv2.cvtColor(imgWhite,cv2.COLOR_BGR2GRAY)
        imgWhite = np.expand_dims(imgWhite,axis=0)
        imgWhite = imgWhite/255
        predictions = model.predict(imgWhite)
        class_label = np.argmax(predictions)
        cv2.putText(ImageShow, f'Class:{folder[class_label]}', (10,30),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
        #cv2.putText(ImageShow, f'Class:{class_label}', (10,30),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
        print(class_label)



    
    #cv2.imshow("image",img)
    cv2.imshow("imageShow",ImageShow)

    key = cv2.waitKey(1)
    if key == ord('s'):
        break
cap.release()
cv2.destroyAllWindows()
```
